At this point I am given a list of 10 tickers that has the same correlation, I have to produce to optimum percentage for them to produce the maximum risk

```python
lstofstocks = ['OXY', 'BIIB', 'SLB', 'GM', 'COP', 'PYPL', 'BA', 'LLY', 'SPG', 'AIG']
```

```python
# Using the 35% distribution 

startdate = '2021-10-01'
enddate= '2021-11-01'


#Determine number of shares bought for each stocks under 35% weighted profolio ($100,000)
# Assuming that the stocks i the list are given in increasing order
NumOfShares = [] # number of shares will be printed in the list according to the corrosponding order


def loopshare(lst, date_start, date_end):
    for i in range (len(lst)):
        if (i < 2):
            NumOfShares.append(35000/((yf.Ticker(lst[i]).history(start=date_start, end=date_end)).iloc[0,3]))
        elif(i < 3):
            NumOfShares.append(12000/((yf.Ticker(lst[i]).history(start=date_start, end=date_end)).iloc[0,3]))
        elif(i < 4):
            NumOfShares.append(3000/((yf.Ticker(lst[i]).history(start=date_start, end=date_end)).iloc[0,3]))
        else:
            NumOfShares.append(2500/((yf.Ticker(lst[i]).history(start=date_start, end=date_end)).iloc[0,3]))
    return NumOfShares
loopshare(lstofstocks, startdate, enddate)
```

```python
listOfClose = [] #list of dataframe

#Extracting the closed data
def loopClose(lst, start_date, end_date):
    for i in range (len(lst)):
        listOfClose.append(pd.DataFrame((yf.Ticker(lst[i]).history(start=start_date, end=end_date))['Close']))
loopClose(lstofstocks, startdate, enddate)
```

```python
listBefore = []

def loopCloseBefore(lst):
    for i in range (len(lst)):
        if (i == 0):
            listBefore.append(lst[i])
        else:
            listBefore.append(lst[i]['Close'])

loopCloseBefore(listOfClose)
```

```python
finalPortfolio = pd.concat(listBefore, join = 'inner', axis=1)

finalPortfolio.columns = lstofstocks
finalPortfolio
```

```python
#Adding portfolio

#function that caculates the porfolio
def portfolio (df):
    count = 0
    for i in range(len(df.columns)):
        count += NumOfShares[i]*(df.columns[i])
    return count

finalPortfolio['portfolio'] = finalPortfolio.OXY* NumOfShares[0]+ finalPortfolio.BIIB*NumOfShares[1] + finalPortfolio.SLB*NumOfShares[2] + finalPortfolio.GM*NumOfShares[3] + finalPortfolio.COP*NumOfShares[4] + finalPortfolio.PYPL*NumOfShares[5] + finalPortfolio.BA*NumOfShares[6] + finalPortfolio.LLY*NumOfShares[7] + finalPortfolio.SPG*NumOfShares[8]+ finalPortfolio.AIG*NumOfShares[9]
finalPortfolio.head()
```
